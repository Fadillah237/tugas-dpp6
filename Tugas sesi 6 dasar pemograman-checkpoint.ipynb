```python
print('======Selamat Datang di Disney Land=====')
umur = int(input('Silahkan masukan umur anda : '))
tinggi = int(input('Silahkan Masukan Tinggi badan anda : '))
tiket = 0
if umur < 2:
    print('============ Dilarang Masuk ============')
elif umur < 4:
    tiket += 30000
    if tinggi > 70:
        tiket += 10000
    print('Harga Tiket = {}'.format(tiket))
elif umur < 7:
    tiket += 40000
    if tinggi > 120:
        tiket += 15000
    print('Harga Tiket = {}'.format(tiket))
elif umur < 10:
    tiket += 50000
    if tinggi > 150:
        tiket += 20000
    print('Harga Tiket = {}'.format(tiket))
else:
    tiket += 80000
    print('Harga Tiket = {}'.format(tiket))
print('==============Terima Kasih==============')
```
